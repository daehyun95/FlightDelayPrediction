````python
import matplotlib.pyplot as plt
from collections import OrderedDict
from mpl_toolkits.basemap import Basemap

airport = "/content/drive/MyDrive/airports.csv"
airport = pd.read_csv(airport)
airport= airport.dropna()

def map_flight(df_ex):
  flightcount = df_ex['Origin'].value_counts()

  plt.figure(figsize=(10,10))

  colors = ['purple', 'green', 'orange','yellow', 'red', 'lightblue']
  size = [1, 100, 1000, 10000, 100000, 1000000]
  labels = ["1 to 100", "100 to 1000", "1000 to 10000", "10000 to 100000", "100000 to 1000000"]

  map = Basemap(llcrnrlon=-180, urcrnrlon=-50, llcrnrlat=10, urcrnrlat=75, lat_0=0, lon_0=0,)
  map.shadedrelief()
  map.drawcoastlines()
  map.drawcountries(linewidth = 4)
  map.drawstates(color='0.3')

  for index, (code, y,x) in airport[['IATA_CODE', 'LATITUDE', 'LONGITUDE']].iterrows():
      x, y = map(x, y)
      if code != "ILG" and code != "CEC" and code != "CLD" and code != "ISN" and code != "MMH":
        isize = [i for i, val in enumerate(size) if val < flightcount[code]]
        ind = isize[-1]
        map.plot(x, y, marker='o', markersize = ind+5, markeredgewidth = 1, color = colors[ind], markeredgecolor='k', label = labels[ind])

  handles, labels = plt.gca().get_legend_handles_labels()
  labels_ord = OrderedDict(zip(labels, handles))
  keys = ('1 to 100','100 to 1000', '1000 to 10000', '10000 to 100000', '100000 to 1000000')
  lnew = OrderedDict()
  for item in keys:
    lnew[item] = labels_ord[item]
  plt.legend(lnew.values(), lnew.keys(), loc = 1, prop= {'size':10}, title='Flights per year', frameon = True, framealpha = 1)
  plt.show()
````

````python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics, linear_model

from sklearn import tree
from sklearn.metrics import confusion_matrix, classification_report
````

````python
df_pp = df_multiple_sample.drop(["FlightDate", "Diverted", "DepartureDelayGroups"], axis=1)
````

````python
df_pp['Cancelled']=df_pp['Cancelled'].astype(int)
````

````python
length_min = df_pp.DepDelayMinutes.max()

minute_dict = {0: 0, 1: 15, 2: 30, 3:60, 4: 1000000000000}
minute_dict
````

````python
DepDelayMin = []

for index, row in df_pp.iterrows():
  i = row.DepDelayMinutes
  for key, value in minute_dict.items():
    if float(i) <= float(value):
      DepDelayMin.append(key)
      break
      

df_pp["DepDelayMin"] = DepDelayMin
df_pp
````

````python
airline = dict()
i = 0
for value in set(df_pp.Airline):
  airline[i] = value
  i+=1

airline
````

````python
Airline = []

for index, row in df_pp.iterrows():
  i = row.Airline
  for key, value in airline.items():
    if (i) == (value):
      Airline.append(key)
      break

df_pp["Airlines"] = Airline
df_pp
````

````python
airport_1 = dict()
i = 0
for value in set(df_pp.Origin):
  airport_1[i] = value
  i+=1

for value in set(df_pp.Dest):
  if value not in airport_1.values():
    airport_1[i] = value
    i+=1

airport_1
````

````python
Origin = []

for index, row in df_pp.iterrows():
  i = row.Origin
  for key, value in airport_1.items():
    if (i) == (value):
      Origin.append(key)
      break

df_pp["Origins"] = Origin
df_pp
````

````python
Destination = []
for index, row in df_pp.iterrows():
  i = row.Dest
  for key, value in airport_1.items():
    if (i) == (value):
      Destination.append(key)
      break


df_pp["Destinations"] = Destination
df_pp
````

````python
df_pp= df_pp.drop(["Airline", "Origin", "Dest"], axis=1)
````

````python
# "DepDelayMinutes"
df_pp= df_pp.drop(["DepTimeBlk", "Cancelled", "AirTime", "Year","DepDel15" ], axis=1)
````

````python
# "DepDelayMinutes"
df_pp= df_pp.drop(["DepDelayMinutes" ], axis=1)
````

````python
df_pp
````

````python
#For the logistic regression model only, we will scale the data to improve performance. The DT is our interpetable model.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()


y_train = df_pp['DepDelayMin'].copy()
X_train = df_pp.drop('DepDelayMin', axis=1).copy()



# # X scale
scaler.fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, train_size=0.7, random_state=42)
````

````python
def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]])
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop([feature_to_encode], axis=1)
    return(res)
````
